```python
df = pandas.read_csv(r"C:\Users\XPS\Downloads\happiness_report\2016.csv")
df.head()
```

```python
type(df)
```

```python
df.groupby("Region").mean()
```

```python
df.plot.scatter(                     # creating a scatterplot of
    x='Family',                      # using "Family" on the horizontal axis
    y='Freedom',                     # using "Freedom" on the vertical axis 
    c='Health (Life Expectancy)',    # color it by "Health (Life Expectancy)"
    cmap='rainbow',                  # using a rainbow color spectrum
    grid=True,                       # plotting the grid underneath
) 
```

```python
ax = df.groupby("Region").mean()['Happiness Score'].plot.bar()  # grouping by region, taking the average and plotting the Happiness Score as a bar plot

ax.figure.savefig('myplot.jpeg', bbox_inches="tight") # saving the file... without cutting out the lables
```
